## Pre-process the '**transaction**' data and create a new data frame of specific columns.

```python
import os
path = "/content/pulse/data/aggregated/transaction/country/india/state/"
ag_state_list=os.listdir(path)
# ag_state_list


trn = {'State': [], 'Year':[], 'Quater': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}

for i in ag_state_list:
  p_i=path+i+'/'
  # print (p_i)
  agg_yr=os.listdir(p_i)
  for j in agg_yr:
    p_j=p_i+j+'/'
    agg_yr_list=os.listdir(p_j)
    for k in agg_yr_list:
      p_k=p_j+k
      Data=open(p_k, 'r')
      D=json.load(Data)
      # print (D)
      for z in D['data']['transactionData']:
        Name = z['name']
        count = z['paymentInstruments'][0]['count']
        amount= z['paymentInstruments'][0]['amount']
        trn['State'].append(i)
        trn['Year'].append(j)
        trn['Quater'].append(int(k.strip('.json')))
        trn['Transaction_type'].append(Name)
        trn['Transaction_count'].append(count)
        trn['Transaction_amount'].append(amount)
df_ag_state_list = pd.DataFrame(trn)


df_ag_state_list
```

# Create SQLite Database with the db name 'ind_state.db'

```python
import sqlite3
import csv
con = sqlite3.connect("ind_state.db")
```

```python
cur = con.cursor()
```

## Create table with the name 'trn_st' using a variable name 'create_table'


```python
create_table = '''CREATE TABLE IF NOT EXISTS trn_st(
				id INTEGER PRIMARY KEY AUTOINCREMENT,
				State TEXT NOT NULL,
				Year INTEGER NOT NULL,
        Quater INTEGER NOT NULL,
        Transaction_count INTEGER NOT NULL,
        Transaction_amount INTEGER NOT NULL);
				'''
```

## Execute the 'create_table' variable using 'cursor' class.

```python
cur.execute(create_table)
```

## Read the .csv data.

```python
file = open('summary_df2.csv')
contents = csv.reader(file)
#
```

## Create a variable to insert the records of each columns in the .csv file.

```python
insert_records = "INSERT INTO trn_st (State, Year, Quater, Transaction_count, Transaction_amount) VALUES(?, ?, ?, ?, ?)"
```

## Insert many records using 'execute_many' function.

```python
cur.executemany(insert_records, contents)
```

## To read all the inserted records in the table; create variable as follows.

```python
select_all = "SELECT * FROM trn_st"
rows = cur.execute(select_all).fetchall()
```

## Call the variable using for-loop.

```python
for r in rows:
	print(r)
```
