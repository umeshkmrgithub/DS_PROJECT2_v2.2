## For loop code to capture the required information from the data and append them in the relevant container.

````
map_trn = os.listdir(path)
map_trn
````

````
maptrndata = {'State': [], 'Year': [], 'Quater':[] ,'District': [], 'Count': [], 'Amount':[]}
map_trn = os.listdir(path)

for i in map_trn:
    p_i = path+i+'/'
    # print (p_i)
    map_yr = os.listdir(p_i)

    for j in map_yr:
        p_j = p_i+j+'/'
        # print (p_j)
        map_qtr = os.listdir(p_j)

        for k in map_qtr:
            p_k = p_j+k
            # print (p_k)
            mapdata = open(p_k, 'r')
            # print (mapdata)
            jmapdata = json.load(mapdata)
            # print (jmapdata)

            if jmapdata ['data']['hoverDataList']is not None:
                for z in jmapdata ['data']['hoverDataList']:
                    district = z['name']
                    count = z['metric'][0]['count']
                    amount = z ['metric'][0]['amount']

                    maptrndata ['State'].append(i)
                    maptrndata ['Year'].append(j)
                    maptrndata ['Quater'].append(int(k.strip('.json')))
                    maptrndata ['District'].append(district)
                    maptrndata ['Count'].append(count)
                    maptrndata ['Amount'].append(amount)

df_maptrndata = pd.DataFrame(maptrndata)
df_maptrndata
````

## Export the required data as .csv files.

````
df3 = df_maptrndata.to_excel('DF3_MAPDATA_DISTRICT.xlsx', index=False)
df3 = df_maptrndata.to_csv('DF3_MAPDATA_DISTRICT.csv', index=False)

````
