```
import pandas as pd
import json
import os
```

```
agusr = {'State': [], 'Year':[], 'Quater': [], 'Brand': [], 'Count': [], 'Percentage': []}
agusr_list = os.listdir(path)

for i in agusr_list:
  p_i=path+i+'/'
  # print (p_i)
  agusr_yr=os.listdir(p_i)
  for j in agusr_yr:
    p_j = p_i+j+'/'
    # print (p_j)
    agusr_yr_list= os.listdir(p_j)
    for k in agusr_yr_list:
      p_k=p_j+k
      # print (p_k)
      Data = open(p_k, 'r')
      D=json.load(Data)
      # print (D)
      if D ['data']['usersByDevice'] is not None:
        for z in D ['data']['usersByDevice']:
          brand = z['brand']
          count = z['count']
          percentage = z['percentage']
          agusr['State'].append(i)
          agusr['Year'].append(j)
          agusr['Quater'].append(int(k.strip('.json')))
          agusr['Brand'].append(brand)
          agusr['Count'].append(count)
          agusr['Percentage'].append(percentage)
df_agusr_list = pd.DataFrame(agusr)
df_agusr_list




```
